```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plot
from sklearn.neural_network import MLPClassifier
```

```python
bc_df = pd.read_csv("ANN_Quiz_Data_Breast_Cancer.csv")
```

```python
from sklearn.model_selection import train_test_split
```

```python
y = bc_df['Target']
x = bc_df.drop(['Target'], axis=1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size= 0.20, random_state=0)
```

```python
# Checking dimensions on the train and test data
print('x_train: ',x_train.shape)
print('x_test: ',x_test.shape)
print('y_train: ',y_train.shape)
print('y_test: ',y_test.shape)
```

```python
clf = MLPClassifier(hidden_layer_sizes=500, activation='relu',
                    verbose=True,  random_state=0,tol=0.0001)
```

```python
clf.fit(x_train, y_train)
y_pred = clf.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
pred2 = clf.predict(x_train)
confusion_matrix(y_train,pred2)
```

```python
clf.score(x_train,y_train)
```

```python
print(classification_report(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt
```

```python
# AUC and ROC for the training data

# predict probabilities
probs = clf.predict_proba(x_train)
# keep probabilities for the positive outcome only
probs = probs[:, 1]
# calculate AUC
from sklearn.metrics import roc_auc_score
auc = roc_auc_score(y_train, probs)
print('AUC: %.3f' % auc)
# calculate roc curve
from sklearn.metrics import roc_curve
fpr, tpr, thresholds = roc_curve(y_train, probs)
plt.plot([0, 1], [0, 1], linestyle='--')
# plot the roc curve for the model
plt.plot(fpr, tpr, marker='.')
# show the plot
plt.show()
```

```python
# AUC and ROC for the test data

# predict probabilities
probs = clf.predict_proba(x_test)
# keep probabilities for the positive outcome only
probs = probs[:, 1]
# calculate AUC
from sklearn.metrics import roc_auc_score
auc = roc_auc_score(y_test, probs)
print('AUC: %.3f' % auc)
# calculate roc curve
from sklearn.metrics import roc_curve
fpr, tpr, thresholds = roc_curve(y_test, probs)
plt.plot([0, 1], [0, 1], linestyle='--')
# plot the roc curve for the model
plt.plot(fpr, tpr, marker='.')
# show the plot
plt.show()
```
