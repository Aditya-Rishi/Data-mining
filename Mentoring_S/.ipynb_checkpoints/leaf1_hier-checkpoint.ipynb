```python
from scipy.stats import zscore
leafs=leaf.apply(zscore)
leafs.head(10)
```

```python
leafs.std()
```

```python
leafs.mean()
```

```python
# Creating the Dendrogram
# Importing dendrogram and linkage module
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
# Choosing average linkage method
```

```python
link_method = linkage(leafs.iloc[:,1:14], method = 'average')
```

```python
labelList = np.array(leafs.Eccentricity)
labelList
```

```python
dend = dendrogram(link_method,labels = labelList)
```

```python
dend = dendrogram(link_method,
                 truncate_mode='lastp',
                 p = 10,
                 )
```

```python
#Importing fcluster module to create clusters
```

```python
from scipy.cluster.hierarchy import fcluster
```

```python
# Set criterion as maxclust,then create 4 clusters, and store the result in another object 'clusters'

#Method 1

clusters = fcluster(link_method, 4, criterion='maxclust')
clusters
```

```python
# Appending clusters to original dataset
```

```python
leaf['clusters'] = clusters
```

```python
leaf.head()
```

```python
# Cluster Frequency
```

```python
leaf.clusters.value_counts().sort_index()
```

```python
# Cluster Profiles
```

```python
aggdata=leaf.iloc[:,1:15].groupby('clusters').mean()
aggdata['Freq']=leaf.clusters.value_counts().sort_index()
aggdata
```

```python
# Cluster 1: Most circular
# Cluster 2: Second most circular
# Cluster 3: Least circular
# Cluster 4: Medium elliptical
```

```python
# Using Agglomerative Clustering
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
# Set n_clusters=4,affinity='euclidean', linkage='average' and store the result in another object 'Cluster_agglo'

cluster = AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='average')
Cluster_agglo=cluster.fit_predict(leafs.iloc[:,1:14])
print(Cluster_agglo)
```

```python
# Appending clusters to original dataset
```

```python
leaf["Agglo_CLusters"]=Cluster_agglo
```

```python
leaf.Agglo_CLusters.value_counts().sort_index()
```

```python
# Agglomeraive Cluster Profiles
```

```python
agglo_data=leaf.drop(["clusters"],axis=1).groupby('Agglo_CLusters').mean()
agglo_data['Freq']=leaf.Agglo_CLusters.value_counts().sort_index()
agglo_data
```

```python
# Cluster 1=most round
# Cluster 2=most round
# Cluster 3=most round
# Cluster 4=most round
```

```python
agglo_data.to_csv('leaf_hc1.csv')
```
